O States of data Brazil 2021 é a maior pesquisa realizada no Brasil e serve para ter um panorama geral do que está ocorrendo na área de dados nacional atualmentem. A pesquisa foi realizada pelo Data Hacker em parceria com a Bain & Company.

## Análise exploratória de dados. 
A análise realizada tem o objetivo de cumprir umas das tarefas do programa de mentoria da Universidade dos Dados, onde deviamos utilizar uma das bibliotecas gráficas aprendidas durante a trilha na análise exploratória.


fonte dos dados: https://www.kaggle.com/datasets/datahackers/state-of-data-2021

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('State of Data 2021 - Dataset - Pgina1.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
datahacker= df[["('P1_b ', 'Genero')","('P1_e ', 'Estado onde mora')","('P1_a ', 'Idade')","('P1_a_a ', 'Faixa idade')","('P2_f ', 'Cargo Atual')","('P2_h ', 'Faixa salarial')", "('P2_g ', 'Nivel')", "('P1_h ', 'Nivel de Ensino')"]]
datahacker.head()
```

```python
datahacker.isnull().sum()
```

```python
datahacker.dropna(inplace=True)
```

```python
datahacker.isnull().sum()
```

```python
datahacker.shape
```

```python
#px.bar()
px.bar(data_frame=datahacker, x="('P1_e ', 'Estado onde mora')", 
title='States of data 2021 - Representação dos estados na pesquisa', 
labels={"('P1_e ', 'Estado onde mora')":'Estados', 'count': 'Total'}, 
height=500,
width=1000)
```

```python
datahacker["('P1_e ', 'Estado onde mora')"].value_counts(normalize=True).to_frame()*100
```

O estado de são paulo é o que representa majoritariamente as respostas da pesquisa do datahacker com uma porcentagem de 40%.

```python
px.histogram(data_frame=datahacker, x="('P1_a ', 'Idade')", nbins=100, opacity=0.8,
 color_discrete_sequence=['indianred'] )
```

Podemos ver que a maioria das respostas foram dadas por pessoa entre 24-29 anos de idade, e pela curva temos uma assimetria positiva com mediana menor que a média.

```python
datahacker["('P1_a ', 'Idade')"].describe()
```

```python
salario = datahacker["('P2_h ', 'Faixa salarial')"].value_counts().to_frame()
salario.columns = ['Faixa Salarial']
salario
```

```python
datahacker["('P2_h ', 'Faixa salarial')"].value_counts(normalize=True)*100
```

20% do pessoal que atua com dados recebe R$ 8.001/mês a R$ 12.000/mês

```python
faixa_salarial = {
                            "Menos de R$ 1.000/mês" : 1,
                            "de R$ 1.001/mês a R$ 2.000/mês" : 2,
                            "de R$ 2.001/mês a R$ 3000/mês" : 3,
                            "de R$ 3.001/mês a R$ 4.000/mês" : 4,
                            "de R$ 4.001/mês a R$ 6.000/mês" : 5,
                            "de R$ 6.001/mês a R$ 8.000/mês" : 6,
                            "de R$ 8.001/mês a R$ 12.000/mês" : 7,
                            "de R$ 12.001/mês a R$ 16.000/mês" : 8,
                            "de R$ 16.001/mês a R$ 20.000/mês" : 9,
                            "de R$ 20.001/mês a R$ 25.000/mês" : 10,
                            "de R$ 25.001/mês a R$ 30.000/mês" : 11,
                            "de R$ 30.001/mês a R$ 40.000/mês" : 12,
                            "Acima de R$ 40.001/mês" : 13
                            }
type(faixa_salarial)
```

```python
datahacker["Faixa salarial"] = datahacker["('P2_h ', 'Faixa salarial')"].map(faixa_salarial)
```

```python
datahacker["('P1_b ', 'Genero')"].value_counts(normalize=True).to_frame()*100
```

```python
px.bar(datahacker, x="('P1_b ', 'Genero')", color="('P1_b ', 'Genero')", 
labels=({"('P1_b ', 'Genero')": 'Genêro'

}))
```

```python
px.box(datahacker,y='Faixa salarial', color="('P1_b ', 'Genero')",
labels=({"('P1_b ', 'Genero')": 'Genêro'

}))
```
